```python
from tensorly.cp_tensor import plot_cp_factors, plot_cp_factors_hist
from tensorly.random import random_cp
```

# Input only Cp tensors

```python
shape = (30, 40, 50)
rank = 4

# one target cp tensor
cp_tensor_1 = random_cp(shape, rank)
cp_tensor_2 = random_cp(shape, rank)

_, _ = plot_cp_factors([cp_tensor_1, cp_tensor_2], print=True)
```

```python
_, _ =plot_cp_factors_hist([cp_tensor_1, cp_tensor_2])
```

# Input with title and mode info

```python
_, _ =plot_cp_factors([cp_tensor_1, cp_tensor_2], title=["cptensor1", "cptensor2"], mode_info=["spatial", "time","hyperspectral"]) 
```

```python
_, _ =plot_cp_factors_hist([cp_tensor_1, cp_tensor_2], title=["cptensor1", "cptensor2"], mode_info=["spatial", "time","hyperspectral"]) 
```

# Input with ranges

```python
_, _ =plot_cp_factors([cp_tensor_1, cp_tensor_2], x_lim=[0, 30])
```

```python
_, _ =plot_cp_factors([cp_tensor_1, cp_tensor_2], y_lim=[0, 0.5])
```

# Warnings

```python
_, _ =plot_cp_factors([cp_tensor_1, cp_tensor_2], mode_info=["spatial","spatial", "time","hyperspectral"]) 
```

```python
_, _ =plot_cp_factors([cp_tensor_1, cp_tensor_2], title=["cp_tensor"]) 
```

# Permute factors

```python
_, _ =plot_cp_factors([cp_tensor_1, cp_tensor_2], permute_factors=True) 
```
